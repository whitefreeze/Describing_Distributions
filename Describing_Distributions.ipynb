# Describing Distributions

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy import stats

# Permanently changes the pandas settings
#pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

## Load Data

```python
# Connect and import data
filename = "Data/medical_dataset.csv"
df = pd.read_csv(filename)
```

```python
display(df.info())
```

```python
df.head()
```

## Describe Distributions: 'VitD_levels'

```python
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature, kde=True, ax=ax,stat='probability')
ax.axvline(mean, ls=':', lw=3, color = 'red',label=f'Mean: {mean:.2f}')
ax.axvline(median, ls='--', lw=3, color = 'green',label=f'Median: {median:.2f}')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{col}')
ax.legend();
```

**Questions:**

Are the values Discrete or Continuous?
* Continuous

Does it have a skew? If so, which direction (+/-)
* It does not visually have a skew.


```python
# What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)
kurt = stats.kurtosis(feature, fisher = False)
kurt
```

What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)

* Mesokurtic (Kurtosis ~3)
* Leptokurtic (Kurtosis >3)
* Platykurtic (Kurtosis < 3)

The kurtosis value is 3.013, so accordingly, the kurtosis type is:
* Mesokurtic (Kurtosis ~3)
